# 1. Periodic Orbit

```python
#Dictionary containing parameters of the model
Nvariables = 6
params = dict(tau_e = 8,
              tau_i = 8,
              tau_se=1,
              tau_si=5,
              nu_e = -5,
              nu_i = -5,
              Delta_e = 1,
              Delta_i = 1,
              Jee = 5,
              Jei = 13,
              Jii = 5,
              Jie = 13,
              Iext_i=0,
              Iext_e=6,
              eps=15)
```

### Compute x0, T

```python
status,x_approx,T = InitCondPOHomogeneous_2(Nvariables,params)
```

```python
t0 = 0
tf = T
#Discretization used for integration
h = 0.001
#Number of points to evaluate the time integration
N = int((tf-t0)/h)
#print(N)
time_eval = np.linspace(t0, tf, N)
#Initial condition
x0 = x_approx
```

```python
sol = solve_ivp(HomogeneousSystem, [t0,tf], x0, t_eval=time_eval, method='RK45', rtol=1e-6, atol=1e-9, args=(Nvariables,params))
    
```

```python
fig1=plt.figure(figsize=(12,7))
ax=plt.axes()

plt.xlabel(r'$Time (ms)$',fontsize=30,fontname='Times New Roman')
plt.ylabel(r'$s_E$',fontsize=30,fontname='Times New Roman')
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
# Create custom legend handles for each row
plt.plot(sol.t,sol.y[2,:],linewidth=2,color='black')
plt.axhline(0.1,color="black", ls="-")
#plt.ylim([-8,8])
```

```python
fig1=plt.figure(figsize=(12,7))
ax=plt.axes()

plt.xlabel(r'$s_I$',fontsize=30,fontname='Times New Roman')
plt.ylabel(r'$s_E$',fontsize=30,fontname='Times New Roman')
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
# Create custom legend handles for each row
plt.plot(sol.y[2,:],sol.y[5,:],linewidth=2,color='black')
plt.axhline(0.1,color="black", ls="-")
#plt.ylim([-8,8])
```

### Check we are on PO

```python
# Define Poincare section
def PoincareEvent(t,x,Nvar,ParametersPop):
    return x[2]-0.1

# Event should be triggered when g(x) == 0 <=> x[1]=0
PoincareEvent.terminal = False  # Stop the integration when the event occurs
PoincareEvent.direction = 1    # Trigger when the second component crosses alpha in ascending direction
```

```python
t0 = 0
tf = 1000
#Discretization used for integration
h = 0.001
#Number of points to evaluate the time integration
N = int((tf-t0)/h)
#print(N)
time_eval = np.linspace(t0, tf, N)
sol = solve_ivp(HomogeneousSystem, [t0,tf], x0, t_eval=time_eval, method='RK45', rtol=1e-6, atol=1e-9, events=[PoincareEvent],args=(Nvariables,params))
  
```

```python
fig1=plt.figure(figsize=(12,7))
ax=plt.axes()

plt.xlabel('Time (ms)',fontsize=30,fontname='Times New Roman')
plt.ylabel(r'$s_E$',fontsize=30,fontname='Times New Roman')
plt.scatter(sol.t_events[0][:],sol.y_events[0][:,2],color='red')
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
# Create custom legend handles for each row
plt.plot(sol.t,sol.y[2,:],linewidth=2,color='black')
plt.axhline(0.1,color="black", ls="-")
#plt.ylim([-8,8])
```

```python
fig1=plt.figure(figsize=(12,7))
ax=plt.axes()

plt.xlabel('Time (ms)',fontsize=30,fontname='Times New Roman')
plt.ylabel(r'$s_E$',fontsize=30,fontname='Times New Roman')
#plt.scatter(sol.t_events[0][:],sol.y_events[0][:,2],color='red')
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
# Create custom legend handles for each row
plt.plot(sol.y[2,:],sol.y[5,:],linewidth=2,color='black')
plt.axhline(0.1,color="black", ls="-")
#plt.ylim([-8,8])
```

### See arbitrary initial condition

```python
sol = solve_ivp(HomogeneousSystem, [t0,tf], np.zeros(Nvariables), t_eval=time_eval, method='RK45', rtol=1e-6, atol=1e-9, events=[PoincareEvent],args=(Nvariables,params))
  
```
